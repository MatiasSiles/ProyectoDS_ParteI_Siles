#📋**Variables**


##🦠DfCovid
*   **ZIP Code**: Los codigos postales de las areas geograficas
*   **Week Number**: Indica el numero de semana del año donde se registran los datos sobre casos y testeos de covid. En total existen en el dataframe unas 53 semanas
*   **Week Start**: Fecha de inicio de cada semana, durante los 7 dias de cada semana, se registran los datos cuantitativos
*   **Week End**: Fecha de finalizacion de la semana, todos los datos recolectados se asignan al numero de semana, quedando organizado los datos en las semanas del año
*   **Cases - Acumulative**: Indica el total de casos de la semana
*   **Test - Acumulative**: Indica el total de testeos semanalmente



##💼DfMBA

*   **Person ID**: Numero univoco de cada profesional
*   **Current Job Title**: Trabajo actual del profesional
*   **Decided to Pursue MBA?**: Informa si el profesional sigue o no el master en administracion empresarial
*   **Has Management Experience**: Expresa si el profesional tiene experiencia en gestion

##🎓DfUni

*   **Vidwan-ID**: Numero univoco del profesor
*   **Name**: Nombre del profesor
*   **Position**: Puesto de trabajo en la universidad
*   **Department**: Departamentos de la universidad
*   **Years of Experience**: Años de experiencia en la universidad

#📚**Librerias**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

#📈**Matplotlib**

```python
# Grafico 1:
mpl.style.use('grayscale')
df1Covid = dfCovid.head(20)
x1 = df1Covid['Week Start']
y1 = df1Covid['Deaths - Cumulative']

plt.subplots(figsize=(15,5))
plt.plot(x1,y1,marker='o',color='black',linewidth=1,label='Muertes Totales')

plt.xlabel('Inicio de Cada Semana')
plt.ylabel('Muertes acumuladas')
plt.title('Muertes acumuladas de las primeras 20 semanas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
#Grafico 2:
mpl.style.use('fivethirtyeight')

df2Covid = dfCovid[['Cases - Cumulative','Tests - Cumulative','Deaths - Cumulative']].iloc[20:30]
df2Covid['Cases - Cumulative'] = df2Covid['Cases - Cumulative'].astype(int)

x2 = df2Covid.columns
y2 = df2Covid.sum()

fig, ax = plt.subplots(figsize=(10,5))
ax.bar(x2,y2, color=['red','blue','purple'],label=['Cases','Tests','Deaths'])

for p in ax.patches:
  ax.annotate(f'{int(p.get_height())}',(p.get_x() + p.get_width() / 2, p.get_height()),
              ha='center', va='bottom', fontsize=10)

ax.set_title('Comparacion por Covid-19 (Semana 20-30)')
ax.set_ylabel('Cantidades Acumuladas')
plt.grid(axis='x')
ax.legend()
plt.show()
```

```python
#Grafico 3:
mpl.style.use('ggplot')
data = dfCovid['Deaths - Cumulative'].iloc[30:35]
labels = dfCovid['Week Number'].iloc[30:35]

fig, ax = plt.subplots(figsize=(10,5))
ax.pie(data,labels=labels)

ax.set_title('Comparacion de Muertes Totales(semana 30-35)')
plt.legend(loc='lower left')
plt.show()
```

#📉**Seaborn**

```python
#Grafico 1:
mpl.style.use('bmh')
plt.subplots(figsize=(20,5))
sns.histplot(data=dfCovid,x='Deaths - Cumulative',color=sns.color_palette("colorblind")[6],
             kde=True,label='Deaths - Cumulative')

plt.title('Histograma de Muertes Acumulativas')
plt.ylabel('Frecuencias')
plt.xlabel('Muertes Acumuladas')
plt.legend()
plt.show()
```

```python
#Grafico 2:
dfCovid4 = dfCovid.head(10)
semanas = ['Semana1','Semana2','Semana3','Semana4','Semana5','Semana6','Semans7','Semana8','Semana9','Semana10']
sns.scatterplot(data=dfCovid4,x='Week Start',y='Cases - Cumulative',s=150,hue=semanas)

plt.title('Casos Acumulados(1eras 10 semanas)')
plt.xlabel('Inicio de Cada Semana')
plt.ylabel('Casos Acumulados')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1)) # Coloca la leyenda afuera: (1,1) = (right,upper)
plt.show()
```

```python
#Grafico 3:
mpl.style.use('ggplot')
df5Covid =  dfCovid[['Cases - Weekly','Tests - Weekly','Deaths - Weekly']].iloc[30:40]

plt.subplots(figsize=(10,10))
sns.boxplot(data=df5Covid)

plt.title('Boxplots de Acumulaciones Semanales entre la semana 30 y 40')
plt.ylabel('Cantidades Acumuladas')
plt.grid()
plt.show()
```
